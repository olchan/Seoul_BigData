```python
# 각 DataFrame의 '행정구역' 집합 추출
sales_regions = set(sales_df['행정구역'])
park_regions = set(park_df['행정구역'])
resident_regions = set(resident_df['행정구역'])
weekday_regions = set(weekday_df['행정구역'])
weekend_regions = set(weekend_df['행정구역'])

# 다른 DataFrame에서 존재하지 않는 행정구역 찾기
sales_only = sales_regions - (park_regions | resident_regions | weekday_regions | weekend_regions)
park_only = park_regions - (sales_regions | resident_regions | weekday_regions | weekend_regions)
resident_only = resident_regions - (sales_regions | park_regions | weekday_regions | weekend_regions)
weekday_only = weekday_regions - (sales_regions | park_regions | resident_regions | weekend_regions)
weekend_only = weekend_regions - (sales_regions | park_regions | resident_regions | weekday_regions)

# 결과 출력
print("sales_df에만 있는 행정구역:", sales_only)
print("park_df에만 있는 행정구역:", park_only)
print("resident_df에만 있는 행정구역:", resident_only)
print("weekday_df에만 있는 행정구역:", weekday_only)
print("weekend_df에만 있는 행정구역:", weekend_only)
```
