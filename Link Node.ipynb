```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler


columns_to_analyze = ['행정동별 거주인구', 'normalized_세대당공원면적', '추정매출액', '개업률', '폐업률']

# 1. 변동계수 계산
def coefficient_of_variation(data):
    return np.std(data) / np.mean(data)

cv_values = {col: coefficient_of_variation(df[col]) for col in columns_to_analyze}

# 2. 정보 이득 계산 (연속형 변수를 위한 근사치)
def information_gain(data):
    _, bins = np.histogram(data, bins='auto')
    hist, _ = np.histogram(data, bins=bins)
    prob = hist / len(data)
    entropy = -np.sum(prob * np.log2(prob + 1e-10))
    return 1 - entropy / np.log2(len(bins) - 1)

ig_values = {col: information_gain(df[col]) for col in columns_to_analyze}
```

```python
# 3. 변동계수와 정보 이득을 결합하여 중요도 점수 계산
importance_scores = {col: cv_values[col] * ig_values[col] for col in columns_to_analyze}

# 4. 쌍대비교 행렬 생성
n = len(columns_to_analyze)
pairwise_matrix = np.ones((n, n))

for i in range(n):
    for j in range(i+1, n):
        col_i = columns_to_analyze[i]
        col_j = columns_to_analyze[j]
        ratio = importance_scores[col_i] / importance_scores[col_j]
        pairwise_matrix[i, j] = min(9, max(1/9, ratio))  # 1/9에서 9 사이의 값으로 제한
        pairwise_matrix[j, i] = 1 / pairwise_matrix[i, j]

# 5. 가중치 계산 (기존 코드 사용)
def calculate_weights(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    max_index = np.argmax(eigenvalues)
    eigenvalues = np.real(eigenvalues[max_index])
    eigenvectors = np.real(eigenvectors[:, max_index])
    return eigenvectors / np.sum(eigenvectors)

weights = calculate_weights(pairwise_matrix)

# 6. 일관성 비율 검증 (기존 코드 사용)
def consistency_ratio(matrix, weights):
    n = len(matrix)
    lambda_max = np.sum(np.dot(matrix, weights) / weights)
    consistency_index = (lambda_max - n) / (n - 1)
    random_index = {3: 0.58, 4: 0.9, 5: 1.12, 6: 1.24, 7: 1.32, 8: 1.41, 9: 1.45, 10: 1.49}
    return consistency_index / random_index[n]

cr = consistency_ratio(pairwise_matrix, weights)
print(f"Consistency Ratio: {cr}")

# 7. Q(x) 지수 계산
scaler = StandardScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(df[columns_to_analyze]), columns=columns_to_analyze)

def calculate_index(row):
    return np.dot(row[columns_to_analyze].values, weights)

df['Q(x)'] = normalized_df.apply(calculate_index, axis=1)

# 결과 출력
print("\n변수별 가중치:")
for col, weight in zip(columns_to_analyze, weights):
    print(f"{col}: {weight:.4f}")
```

```python
def optimize_linear_park(G, polygon, Q_data_final):
    model = pulp.LpProblem(f"Linear_Park_Optimization_Polygon_{polygon}", pulp.LpMinimize)

    # 결정 변수
    node_vars = pulp.LpVariable.dicts("Node", G.nodes(), cat='Binary')
    edge_vars = pulp.LpVariable.dicts("Edge", G.edges(), cat='Binary')

    # 목적 함수: 선택된 엣지의 가중치 합 최소화
    edge_weights = pulp.lpSum([edge_vars[e] * G[e[0]][e[1]]['weight'] for e in G.edges()])
    model += edge_weights

    # 제약 조건
    # 1. 정확히 3개의 노드 선택
    model += pulp.lpSum(node_vars.values()) == 3

    # 2. 최소 하나의 최외곽 지역 포함
    model += pulp.lpSum([node_vars[n] for n in G.nodes() if G.nodes[n]['is_outer']]) >= 1

    # 3. 선택된 노드는 연결되어야 함
    for e in G.edges():
        model += edge_vars[e] <= node_vars[e[0]]
        model += edge_vars[e] <= node_vars[e[1]]

    # 4. 선택된 노드들은 경로를 형성해야 함 (인접-최외곽-인접)
    model += pulp.lpSum(edge_vars.values()) == 2

    # 5. Q(x) 값 고려
    outer_nodes = [n for n in G.nodes() if G.nodes[n]['is_outer']]
    adjacent_nodes = [n for n in G.nodes() if not G.nodes[n]['is_outer']]

    outer_q = sum([Q_data_final.loc[Q_data_final['행정구역'] == n, 'Q(x)'].values[0] for n in outer_nodes if n in Q_data_final['행정구역'].values])
    adjacent_q = sum([Q_data_final.loc[Q_data_final['행정구역'] == n, 'Q(x)'].values[0] for n in adjacent_nodes if n in Q_data_final['행정구역'].values])

    if adjacent_q != 0:
        q_ratio = outer_q / adjacent_q
    else:
        q_ratio = 1  # 0으로 나누는 것을 방지

    # Q(x) 비율을 목적 함수에 추가 (최대화하려는 값이므로 음수를 사용)
    q_objective = -q_ratio * 0.01  # 가중치를 조정하여 INTER 값과 균형을 맞춤
    model += q_objective

    # 모델 풀기
    model.solve()

    # 결과 추출
    selected_nodes = [n for n in G.nodes() if node_vars[n].value() > 0.5]
    selected_edges = [e for e in G.edges() if edge_vars[e].value() > 0.5]

    # 목적 함수 값 계산
    edge_weight_sum = sum(G[e[0]][e[1]]['weight'] for e in selected_edges)
    q_value = q_ratio * 0.01 if selected_nodes else 0
    objective_value = edge_weight_sum - q_value

    return selected_nodes, selected_edges, objective_value

# 각 폴리곤에 대해 최적화 실행
results = {}
for polygon in range(1, 7):
    polygon_nodes = set(poly_near[poly_near['폴리곤 번호'] == polygon]['폴리곤 최외곽 행정동'].tolist() +
                        poly_near[poly_near['폴리곤 번호'] == polygon]['폴리곤 인접행정동'].tolist())
    polygon_graph = G.subgraph(polygon_nodes)

    nodes, edges, objective_value = optimize_linear_park(polygon_graph, polygon, Q_data_final)
    results[polygon] = (nodes, edges, objective_value)

# 결과 출력 및 데이터프레임 생성
optimal_paths = []
for polygon, (nodes, edges, objective_value) in results.items():
    print(f"폴리곤 {polygon}:")
    print("선택된 노드:", nodes)
    print("선택된 엣지:", edges)
    print("목적 함수 값:", objective_value)
    print()

    optimal_paths.append({
        'Polygon': polygon,
        'Selected_Nodes': ', '.join(nodes),
        'Selected_Edges': ', '.join([f"({e[0]}-{e[1]})" for e in edges]),
        'Objective_Value': objective_value
    })

optimal_paths_df = pd.DataFrame(optimal_paths)
print("최적 경로 정보:")
print(optimal_paths_df)
```

```python
# 데이터 정제 함수
def clean_district_names(name):
    return [district.strip() for district in name.split(',')]

# poly_near 데이터프레임 정제
poly_near['폴리곤 최외곽 행정동'] = poly_near['폴리곤 최외곽 행정동'].apply(clean_district_names)
poly_near['폴리곤 인접행정동'] = poly_near['폴리곤 인접행정동'].apply(clean_district_names)
poly_near = poly_near.explode('폴리곤 최외곽 행정동').explode('폴리곤 인접행정동')

# 그래프 생성
G = nx.Graph()

# 노드 추가
for _, row in inter_score.iterrows():
    outer_q = Q_data_final[Q_data_final['행정구역'] == row['최외곽 행정동']]['Q(x)'].values[0] if not Q_data_final[Q_data_final['행정구역'] == row['최외곽 행정동']].empty else 0
    adjacent_q = Q_data_final[Q_data_final['행정구역'] == row['인접 행정동']]['Q(x)'].values[0] if not Q_data_final[Q_data_final['행정구역'] == row['인접 행정동']].empty else 0

    G.add_node(row['최외곽 행정동'], q=outer_q, is_outer=True)
    G.add_node(row['인접 행정동'], q=adjacent_q, is_outer=False)

# 엣지 추가
for _, row in inter_score.iterrows():
    G.add_edge(row['최외곽 행정동'], row['인접 행정동'], weight=row['INTER'])
def optimize_linear_park_exact(G, polygon):
    # 그래프에 노드가 3개 미만인 경우 처리
    if len(G.nodes()) < 3:
        print(f"폴리곤 {polygon}에는 노드가 3개 미만입니다.")
        return [], []

    # 모든 조합 생성
    node_combinations = list(itertools.combinations(G.nodes(), 3))
    best_combination = None
    best_score = float('inf')

    for combination in node_combinations:
        # 조합의 엣지 찾기
        edges = [(combination[0], combination[1]), (combination[1], combination[2]), (combination[0], combination[2])]
        edge_weights = [G[edge[0]][edge[1]]['weight'] for edge in edges if G.has_edge(edge[0], edge[1])]

        # 모든 엣지가 존재해야만 평가
        if len(edge_weights) == len(edges):
            score = sum(edge_weights)  # 예를 들어, 가중치 합계로 최적화
            if score < best_score:
                best_score = score
                best_combination = combination

    return best_combination, [(best_combination[0], best_combination[1]), (best_combination[1], best_combination[2]), (best_combination[0], best_combination[2])] if best_combination else []

# 각 폴리곤에 대해 최적화 실행
results = {}
for polygon in range(1, 7):
    polygon_nodes = set(poly_near[poly_near['폴리곤 번호'] == polygon]['폴리곤 최외곽 행정동'].tolist() +
                        poly_near[poly_near['폴리곤 번호'] == polygon]['폴리곤 인접행정동'].tolist())
    polygon_graph = G.subgraph(polygon_nodes)

    print(f"폴리곤 {polygon} 노드 수: {len(polygon_graph.nodes())}")
    print(f"폴리곤 {polygon} 엣지 수: {len(polygon_graph.edges())}")

    nodes, edges = optimize_linear_park_exact(polygon_graph, polygon)
    results[polygon] = (nodes, edges if edges else [])

# 결과 출력 및 CSV 저장
output_data = []
for polygon, (nodes, edges) in results.items():
    print(f"폴리곤 {polygon}, 선택된 노드:", nodes)
    print("선택된 엣지:", edges)
    output_data.append((polygon, nodes, edges))

output_df = pd.DataFrame(output_data, columns=['폴리곤 번호', '선택된 노드', '선택된 엣지'])
output_df.to_csv('top3_optimal.csv', index=False)
```

```python
import pandas as pd
import geopandas as gpd
import folium
from shapely.wkt import loads
from shapely.geometry import MultiPolygon, Point

# 올바른 형식으로 변환하는 함수
def convert_to_multipolygon(coord_str):
    if not coord_str or pd.isna(coord_str):  # 비어있거나 NaN인 경우
        return None

    coords = coord_str.split(';')

    # 좌표 형식이 유효한지 확인
    valid_coords = []
    for coord in coords:
        try:
            x, y = coord.split(',')
            valid_coords.append(f"{x.strip()} {y.strip()}")
        except ValueError:
            continue  # 잘못된 좌표는 무시

    if len(valid_coords) < 3:  # 최소 3개 이상의 좌표가 필요
        return None

    coords_str = ', '.join(valid_coords)
    multipolygon_str = f'MULTIPOLYGON(({coords_str}))'

    return loads(multipolygon_str)  # 문자열을 실제 geometry로 변환

# 데이터프레임을 GeoDataFrame으로 변환
poly_near['폴리곤 좌표'] = poly_near['폴리곤 좌표'].apply(convert_to_multipolygon)
poly_near = poly_near[poly_near['폴리곤 좌표'].notna()]  # 유효한 좌표만 필터링
gdf_poly = gpd.GeoDataFrame(poly_near, geometry='폴리곤 좌표').set_crs(epsg=4326)

# 행정구역 GeoDataFrame 생성
gdf_q_data = gpd.GeoDataFrame(q_data, geometry='geometry_y')  # geometry_y를 직접 사용

# 지도 생성
m = folium.Map(location=[37.5, 127.0], zoom_start=12)

# 폴리곤 테두리 추가
# 폴리곤 좌표 시각화 (초록색 테두리)
for idx, row in poly_near.iterrows():
    polygon_coords = row['폴리곤 좌표'].strip("()").split(';')
    polygon_coords = [(float(lon), float(lat)) for lon, lat in (coord.split(',') for coord in polygon_coords)]
    folium.Polygon(locations=polygon_coords, color='green', fill=False).add_to(m)


# 최적 경로 시각화
for idx, row in optimal_paths_df.iterrows():
    selected_nodes = row['Selected_Nodes'].strip("[]").replace("'", "").split(', ')
    selected_edges = row['Selected_Edges'].strip("()").replace("'", "").split(', ')

    # 노드 경로 추가 (빨간색)
    node_coords = []
    for node in selected_nodes:
        geom = gdf_q_data[gdf_q_data['행정구역'] == node]['geometry_y']
        if not geom.empty:
            node_coords.append(geom.values[0].representative_point())  # 중심점 사용

    folium.PolyLine(locations=[(pt.y, pt.x) for pt in node_coords], color='red', weight=3).add_to(m)

    # 엣지 경로 추가 (파란색)
    for edge in selected_edges:
        start_node, end_node = edge.split('-')

        start_geom = gdf_q_data[gdf_q_data['행정구역'] == start_node]['geometry_y']
        end_geom = gdf_q_data[gdf_q_data['행정구역'] == end_node]['geometry_y']

        if not start_geom.empty and not end_geom.empty:
            edge_coords = [
                (start_geom.values[0].representative_point().y, start_geom.values[0].representative_point().x),
                (end_geom.values[0].representative_point().y, end_geom.values[0].representative_point().x)
            ]
            folium.PolyLine(locations=edge_coords, color='blue', weight=2).add_to(m)
        else:
            print(f"경고: '{start_node}' 또는 '{end_node}'에 대한 경계가 없습니다.")

# HTML 파일로 저장
m.save('optimized_route_map.html')
```
